## Setup a simple server

Trong thư mục root của project ta tạo một file có tên là server.py có nội dung như sau:

```python
from fastapi import FastAPI # import class FastAPI từ thư viện fastapi

app = FastAPI() # Khai báo một server app

@app.get("/") # khai báo một root route với Method GET
async def root():
  return "Hello world"
```

Để start server, Chạy câu lệnh sau trên command line:

> uvicorn server:app

## Query Parameters - GET

Để nhận dữ liệu được gửi lên bởi phương thức GET thì ta truyền param dưới dạn key-value vào hàm.

Ví dụ ta có một request như sau: `http://127.0.0.1:8000/items?index=1&step=1`

```python
from fastapi import FastAPI

app = FastAPI()

items_db = [{"item_name": "Foo"}, {"item_name": "Bar"}, {"item_name": "Baz"}]

@app.get("/items/")
async def get_item(index: int = 0, step: int = 10):
  return items_db[index: index+step]
```

## Request Body - POST

Đối với phương thức post ta có chút khác biệt, vẫn ví dụ trên ta có thể thực hiện như sau.

Nhưng trước khi vào code ta phải cài đặt thư viện `pydantic`

```python
!pip install pydantic
```

Và đây là code ví dụ:

```python
from fastapi import FastAPI
from pydantic import BaseModel

app = FastAPI()

items_db = [{"item_name": "Foo"}, {"item_name": "Bar"}, {"item_name": "Baz"}]

# Khai báo class kế thừa BaseModel, đóng vai trò là param cho request body
class Item(BaseModel):
  index: int
  step: int

@app.post("/items/")
async def get_item(item: Item):
  return items_db[item.index: item.index+item.step]
```
